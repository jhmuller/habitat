```python
import os
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import datetime

from IPython.display import display
```

```python
# read the file
file_path = os.path.join(".", "sip_dam_data.csv")
df = pd.read_csv(file_path, index_col=0)
display(df.head(2))
```

```python
# index looks like a datetime, so convert
df.index.name = "dt"
df.index = pd.to_datetime(df.index)
df.sort_index(inplace=True)
```

```python
# check for nulls
nulldf = pd.DataFrame(df.isnull().sum())
nulldf.columns = ["#nulls"]
display(nulldf)
# can use the slices below to see where the nulls are
# they contiguous for both but on diffeent days
sipnull = df.loc[df["SIP"].isnull()]
damnull = df.loc[df["DAM"].isnull()]
```

```python
# basic stats on the non null data
display(df.agg([np.min, np.mean, np.std, np.max]))
```

```python
# plots to explore patterns
fig, axs = plt.subplots(nrows=2, ncols=1, figsize=[16, 8],)
pcols = ["SIP", "DAM"]
# use the start_dt and numdays to explore different
# parts of the series
# or set start_dt to null to see the full plot
start_dt = datetime.datetime(2021, 2, 1) 
#start_dt = None
if start_dt:
    numdays = 20
    xtick_freq = 1 # in days
    xticks = [start_dt + datetime.timedelta(days=x) for x in range(1, numdays, xtick_freq)]
    end_dt = max(xticks)
for i, (ax, col) in enumerate(zip(axs, pcols)):
    if start_dt:
        ax = axs[i]
        ax.set_xlim(start_dt, end_dt)
        ax.set_xticks(xticks,)
        ylim = df.loc[(df.index >= start_dt) & (df.index <= end_dt), col].agg([min, max])  
        ax.set_ylim([ylim[0]-2, ylim[1]+2])        
    ax.plot(df.index, df[pcols[i]],)
    ax.set_title(col)
    ax.grid()
```

```python
# Check to see if there are patterns in timeofday and dayofyear
df["dt"] = df.index
df["dayofyear"] = df["dt"].apply(lambda x: (x - datetime.datetime(x.year, 1, 1)).days)
df["timeofday"] = df["dt"].apply(lambda x: (x - datetime.datetime(x.year, x.month, x.day)))
for window in ["dayofyear", "timeofday"]:
    df[[window, "SIP"]].groupby(window).agg([np.mean, np.median]).plot(figsize=[10, 4], title=window)
    df[[window, "DAM"]].groupby(window).agg([np.mean, np.median]).plot(figsize=[10, 4], title=window)
```

```python
# Look at 2 standard time series metrics
#  ACF and PACF

import statsmodels.tsa.api as tsa
pcols = ["SIP","DAM"]
fig, axs = plt.subplots(nrows=2, ncols=1, figsize=[16, 8],)
for i, (ax, col) in enumerate(zip(axs, pcols)):
    notnull = df.loc[df[col].notnull()]
    acf = tsa.stattools.acf(notnull[col], )
    pacf = tsa.stattools.pacf(notnull[col])
    ax.plot(acf, color="red", label="acf")
    ax.plot(pacf, color="blue", label="pacf")
    ax.set_title(col)
    ax.grid()
    ax.legend()
```

Observations about the series

* There are short periods of missing data in both series
* Clear daily patterns in both SIP and DAM, more regular in DAM
    - 2 humps or spikes, one around 9 AM another around 5 PM
* Appears to be an annual pattern as well.  
    This seems odd as I would expect spring and fall to be similar
* The spikes in one often accompanied by spikes in the other
    - plenty of exceptions e.g. May 2017 spike in SIP
* I suspect there is some useful information from one series
    that can be used to predict the other, although not obvious
    and as noted above sometimes the relation is absent


Do a few tests to see what the error would
be for two trivial models to forecast DAM
* using the overal mean
* using the time of day mean

Note: I am aware that there is lookahead bias in these very simple
   estimates since I am using the whole series to get the means
   rather than using the means up to the point.
   With sufficient history I would think the difference is negligable.

```python
def rmse(x, y):
    return np.round((np.nanmean((x-y)**2))**.5, 3)

def mad(x, y):
    return np.round(np.nanmean(np.abs(x-y)), 3)

cutoff = datetime.datetime(2020,1,1)
before = df.loc[df.index < cutoff, "DAM"]
after = df.loc[df.index >= cutoff, "DAM"]
meandam = before.mean()
print(f'Predicting using overall mean:\n rmse {rmse(after, meandam)}  mad {mad(after, meandam)}\n')


tod = df[["timeofday", "DAM"]].groupby("timeofday").agg(np.median)
tod = np.resize(tod.values, after.shape[0])
print(f'Predicting using time of day mean:\n rmse {rmse(after.values, tod)} mad {mad(after.values, tod)}\n')
```

```python
mm = df.copy()
for col in ["DAM", "SIP"]:
    mm[f"prevMin{col}"] = mm[col].rolling(48).min()
    mm[f"prevMax{col}"] = mm[col].rolling(48).max()
    mm[f"prevSum{col}"] = mm[col].rolling(48).sum()
mm["tod"] = mm["timeofday"].apply(lambda x: str(x)[7:])
tod = pd.get_dummies(mm.tod, prefix='TOD')
mm = mm.join(tod)
mm.tail()
```

```python
import sklearn
from sklearn.ensemble import RandomForestRegressor
from sklearn.datasets import make_regression
ycol = "DAM"
xcols = [c for c in mm.columns if c.startswith("TOD")] #+ ["prevMinDAM", "prevMaxDAM", "prevSumDAM", "prevMinSIP", "prevMaxSIP"]
train_cutoff = datetime.datetime(2019,1,1)
mm.dropna(how='any',axis=0, inplace=True)
X_train = mm.loc[(mm.index < train_cutoff), xcols]
y_train = mm.loc[(mm.index < train_cutoff), "DAM"]
regr = RandomForestRegressor(max_depth=2, random_state=0)
regr.fit(X_train, y_train)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression(fit_intercept=False)
reg.fit(X_train, y_train)
X_train.head(2)
```

```python
coeffs = pd.Series(reg.coef_, index=xcols)
print(reg.coef_)
coeffs.plot.bar( figsize=[16, 4], grid=True)
```

```python
for end in [datetime.datetime(2020, 1, 1), datetime.datetime(2021, 1, 1)]:
    X_test = mm.loc[(mm.index >= train_cutoff) & (mm.index < end), xcols]
    y_test = mm.loc[(mm.index >= train_cutoff)& (mm.index < end), "DAM"]
    y_pred = reg.predict(X_test)
    print(rmse(y_test, y_pred), mad(y_test, y_pred))
```
